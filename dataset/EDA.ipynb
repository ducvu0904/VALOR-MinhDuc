```python
from generate import create_synthetic_data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = create_synthetic_data()
display(df.head())
```

```python
label = df['label']

print(label.describe())
print("Zero fraction:", (label == 0).mean())
print("Non-zero fraction:", (label > 0).mean())

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(label, bins=50)
axes[0].set_title("Label histogram (raw)")
axes[0].set_yscale("log")

positive = label[label > 0]
axes[1].hist(np.log1p(positive), bins=50)
axes[1].set_title("Log1p label (positive only)")

plt.tight_layout()
plt.show()
```
